```python
import pandas as pd
```

```python
# On passe en latin1 pour gérer les accents français (é, à, è...)
radar_df = pd.read_csv(
    "jeu-de-donnees-liste-des-radars-fixes-en-france-12-2025.csv", 
    encoding="latin1", 
    sep=";"
)

# On vérifie si les troupes sont en place
radar_df
```

```python
print(radar_df.columns.tolist())
```

```python
radar_df.columns = radar_df.columns.str.lower()

radar_df = radar_df.rename(columns={'numéro de radar': 'numero de radar'})
radar_df.columns = radar_df.columns.str.strip().str.replace(' ', '_')
radar_df['date_de_mise_en_service'] = pd.to_datetime(radar_df['date_de_mise_en_service']).dt.strftime('%Y-%m-%d')
radar_df['vma'] = radar_df['vma'].fillna('inconnu')

radar_df
```

```python
radar_df.to_csv('radar_clean', index=False, sep=';', encoding='utf-8-sig')
```
