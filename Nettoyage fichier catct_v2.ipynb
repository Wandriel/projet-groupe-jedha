```python
import pandas as pd 
import numpy as np
```

```python
df_2021 = pd.read_csv('Dataset/Dataset - Sécurité Routière.xlsx - caract 2021.csv', sep=';')
df_2022 = pd.read_csv('Dataset/Dataset - Sécurité Routière.xlsx - caract 2022.csv', sep=';')
df_2023 = pd.read_csv('Dataset/Dataset - Sécurité Routière.xlsx - caract 2023.csv', sep=';')
df_2024 = pd.read_csv('Dataset/Dataset - Sécurité Routière.xlsx - caract 2024.csv', sep=',')

# Now concatenate
caract_df = pd.concat([df_2024, df_2023, df_2022, df_2021], ignore_index=True)

print(caract_df.head())
```

```python
caract_df.info()
```

```python
caract_df.isnull().sum()
```

```python
# On travaille uniquement sur le DataFrame global 'caract_df'
h_fix = caract_df['hrmn'].astype(str).str.replace(':', '').str.replace('nan', '0000').str.zfill(4)

date_string = (
    caract_df['an'].astype(str) + '-' + 
    caract_df['mois'].astype(str) + '-' + 
    caract_df['jour'].astype(str) + ' ' + 
    h_fix.str[:2] + ':' + h_fix.str[2:]
)

caract_df['date'] = pd.to_datetime(date_string, errors='coerce')

# Vérification
print(f"Nombre de dates invalides : {caract_df['date'].isna().sum()}")
```

```python
caract_df.isnull().sum()
```

```python
caract_df.info()
```

```python
caract_df = caract_df.drop(columns=['Accident_Id'])
```

```python
caract_df = caract_df.drop(columns=['an', 'mois', 'jour', 'adr'])
print(caract_df.info())
```

```python
caract_df
```

```python
nouveaux_noms = {
    'Num_ACC': 'luminosite',
    'dep': 'departement',
    'com': 'commune',
    'agg': 'agglomeration',
    'int': 'intersection',
    'atm': 'meteo',
    'col': 'type_collision'
}

# On applique le changement
caract_df = caract_df.rename(columns=nouveaux_noms)

# On vérifie le résultat
print(caract_df.columns)
caract_df.head()
```

```python
colonnes_a_traiter = ['meteo', 'type_collision']
caract_df[colonnes_a_traiter] = caract_df[colonnes_a_traiter].replace(-1, np.nan)
caract_df
```
