```python
# Chargement des deux premiers
df2021 = pd.read_csv('projet-groupe-jedha/Dataset/Dataset - Sécurité Routière.xlsx - vehicules 2021.csv', sep=';', encoding='latin-1', dtype=str)
df2022 = pd.read_csv('projet-groupe-jedha/Dataset/Dataset - Sécurité Routière.xlsx - vehicules 2022.csv', sep=';', encoding='latin-1', dtype=str)

# Nettoyage des noms de colonnes
df2021.columns = df2021.columns.str.strip().str.replace('"', '')
df2022.columns = df2022.columns.str.strip().str.replace('"', '')

# Tentative de fusion
fusion_1 = pd.concat([df2021, df2022], ignore_index=True)

print("--- TEST 1 : 2021 + 2022 ---")
print(f"Nombre de colonnes : {len(fusion_1.columns)}")
print(f"Colonnes : {list(fusion_1.columns)}")
```

```python
df2023 = pd.read_csv('projet-groupe-jedha/Dataset/Dataset - Sécurité Routière.xlsx - vehicules 2023.csv', sep=';', encoding='latin-1', dtype=str)
df2023.columns = df2023.columns.str.strip().str.replace('"', '')

# On ajoute 2023 à notre première fusion
fusion_2 = pd.concat([fusion_1, df2023], ignore_index=True)

print("--- TEST 2 : Ajout de 2023 ---")
print(f"Nombre de colonnes : {len(fusion_2.columns)}")
print(f"Colonnes : {list(fusion_2.columns)}")
```

```python
# On utilise la virgule ici comme tu l'as identifié
df2024 = pd.read_csv('projet-groupe-jedha/Dataset/Dataset - Sécurité Routière.xlsx - vehicules 2024.csv', sep=',', encoding='latin-1', dtype=str)
df2024.columns = df2024.columns.str.strip().str.replace('"', '')

# On ajoute 2024 à la fusion précédente
fusion_finale = pd.concat([fusion_2, df2024], ignore_index=True)

print("--- TEST 3 : Ajout de 2024 ---")
print(f"Nombre de colonnes : {len(fusion_finale.columns)}")
print(f"Colonnes : {list(fusion_finale.columns)}")
```

```python
# Sauvegarde du résultat final avec le point-virgule comme standard
fusion_finale.to_csv('fusion_vehicules_21_24_OK.csv', index=False, sep=';')

print("Fichier sauvegardé avec succès sous le nom : fusion_vehicules_21_24_OK.csv")
print(f"Total final : {len(fusion_finale)} lignes et {len(fusion_finale.columns)} colonnes.")
```

# nettoyage de immatriculation. Il n'a pas été push pour le moment.

```python


fichiers_verif = {
    '2021': (';', 'projet-groupe-jedha/Dataset/Dataset - Sécurité Routière.xlsx - véhi imma 2021.csv'),
    '2022': (';', 'projet-groupe-jedha/Dataset/Dataset - Sécurité Routière.xlsx - véhi imma 2022.csv'),
    '2023': (';', 'projet-groupe-jedha/Dataset/Dataset - Sécurité Routière.xlsx - véhi imma 2023.csv'),
    '2024': (',', 'projet-groupe-jedha/Dataset/Dataset - Sécurité Routière.xlsx - véhi imma 2024.csv')
}

for annee, (sep, chemin) in fichiers_verif.items():
    df_test = pd.read_csv(chemin, sep=sep, encoding='latin-1', nrows=1)
    print(f"\n--- ANNÉE {annee} ---")
    for i, col in enumerate(df_test.columns):
        print(f"Index {i} : {col}")
```

```python
# REMPLACE LES CHIFFRES PAR CEUX QUE TU AS VUS AU DESSUS
config_reelle = {
    '2021': (';', 7), # Si c'est bien 7
    '2022': (';', 6), # À vérifier
    '2023': (';', 6), # À vérifier (L'erreur venait sûrement d'ici !)
    '2024': (',', 6)  # À vérifier
}

liste_imma_clean = []

for annee, (sep, idx_age) in config_reelle.items():
    chemin = fichiers_verif[annee][1]
    df_temp = pd.read_csv(chemin, sep=sep, encoding='latin-1', dtype=str)
    
    df_clean = pd.DataFrame()
    df_clean['Num_Acc'] = df_temp.iloc[:, 0]
    df_clean['Age_vehicule'] = df_temp.iloc[:, idx_age]
    df_clean['Num_Acc'] = df_clean['Num_Acc'].str.replace(r'[^\d]', '', regex=True)
    
    liste_imma_clean.append(df_clean)
    print(f"✅ Année {annee} ok.")

df_age_total = pd.concat(liste_imma_clean, ignore_index=True).drop_duplicates()
print("\nFusion des âges terminée !")
```
